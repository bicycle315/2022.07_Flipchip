```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0.0000000, 0.000005, 100)
arccosh = np.arccosh(x**(-1))
plt.plot(x, arccosh)
y=x**(-1)
plt.plot(x, y)
```

```python
t=np.arccosh(x)
plt.plot(x,t)
```
